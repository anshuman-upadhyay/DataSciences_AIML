Creating test set

using random shuffling

```python
def shuffle_and_split(data,test_ratio):
    np.random.seed(23)
    shuffeled_index=np.random.permutation(len(data))# this returns shuffeled data
    test_set_size=int(len(data)* test_ratio)
    test_index= shuffeled_index[:test_set_size]
    train_index= shuffeled_index[test_set_size:]
    return data.iloc[train_index],data.iloc[test_index]
```

using stratified shuffle split 

```python
 # Separating median income into different parts 
df["income_cat"]=pd.cut(df["median_income"],bins=[0,1.5,3.0,4.5,6.0,np.inf],labels=[1,2,3,4,5])
df.head()
```

```python
df["income_cat"].value_counts().sort_index().plot.bar(rot=0,grid=True)
plt.title("Income Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of instances ")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)

for train_index,test_index in split.split(df,df["income_cat"]):
    train_set=df.iloc[train_index]
    test_set=df.iloc[test_index]
```

```python
train_set["income_cat"].value_counts().sort_index().plot.bar(rot=0,grid=True)
plt.title("Income Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of instances ")
```

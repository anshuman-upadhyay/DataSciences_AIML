```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
df=pd.read_csv("housing.csv")
df
```

Data loading and EDA

```python
df.info()
# meadian house value is our lable
```

using stratified shuffle split 

```python
 # Separating median income into different parts 
df["income_cat"]=pd.cut(df["median_income"],bins=[0,1.5,3.0,4.5,6.0,np.inf],labels=[1,2,3,4,5])
```

```python
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)

for train_index,test_index in split.split(df,df["income_cat"]):
    train_set=df.iloc[train_index]
    test_set=df.iloc[test_index]
```

```python
#Removing the income cat column
for sett in(train_set,test_set):
    sett.drop("income_cat",axis=1,inplace=True)
```

```python
df=train_set.copy()
df
```

```python
df.plot(kind="scatter",x="latitude",y="longitude",grid=True,alpha=0.8,cmap="jet",c="median_house_value")
```

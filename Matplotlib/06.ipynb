```python
import matplotlib.pyplot as plt
```

```python
study=[1,2,3,4,5,6,7,8,9]
mark=[40,50,55,60,23,56,77,88,77]
plt.scatter(study,mark)
plt.title("Parai : Marks ratio")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
```

```python
size=[score*2 for score in mark]
colors=["red" if m <60 else "green" for m in mark]
plt.scatter(study,mark,s=size,c=colors)
plt.title("Parai : Marks ratio")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
```

```python
import numpy as np


normal_score=np.array(mark)
sscore=[120,34,554,365,336,345,34,3,343]
plt.scatter(study, mark, c=normal_score,s=sscore,cmap='viridis')
plt.colorbar(label='Score')

for i in range(len(study)):
    plt.annotate(f"S {i+1}", {study[i],mark[i]-2.5})

plt.title('Scatter Plot with Colormap')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
